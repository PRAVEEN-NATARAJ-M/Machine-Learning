```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.barplot(x='total_bill',y='tip',data=tips)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.boxplot(x='total_bill',y='tip',data=tips)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.violinplot(x='total_bill',y='tip',data=tips)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
iris=sns.load_dataset('iris')
```

```python
correlation_matrix=iris.drop('species', axis=1).corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix of iris dataset')
plt.show()
```

```python
sns.pairplot(iris)
plt.title('Pairplot of iris dataset')
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
plt.title('Jointplot of iris dataset')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```
