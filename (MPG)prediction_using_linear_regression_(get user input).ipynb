```python
ind=mpgdf[['cylinders','displacement','horsepower','weight','acceleration']]
dep=mpgdf['mpg']
```

```python
lr=LinearRegression()
lr.fit(ind,dep)
```

```python
c=int(input("ENTER NUMBER OF CYLINDERS:"))
d=int(input("ENTER DISPLACEMENT:"))
h=int(input("ENTER HORSEPOWER:"))
w=int(input("ENTER WEIGHT:"))
a=float(input("ENTER ACCELERATION:"))
m=lr.predict([[c,d,h,w,a]])
print(f"THE MPG(miles per gallon) FOR GIVEN DETAILS IS",m)
```
