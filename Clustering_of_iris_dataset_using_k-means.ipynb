```python
x=iris.iloc[:,[0,1,2,3]].values
print(x) #or x
```

```python
WCSS=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)#max_iter=300,n_init=10,
  kmeans.fit(x)
  WCSS.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),WCSS)
plt.title('THE ELBOW METHOD')
plt.xlabel('NUMBER OF CLUSTERS')
plt.ylabel('WCSS')
plt.show
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
y=kmeans.fit_predict(x)
print(y)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x[y==0,0],x[y==0,1] ,s=100,c='blue',label='Iris-setosa')
plt.scatter(x[y==1,0],x[y==1,1] ,s=100,c='orange',label='Iris-versicolor')
plt.scatter(x[y==2,0],x[y==2,1] ,s=100,c='green',label='Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s=100,c='red',label='Centroids')
plt.legend()
plt.title('CLUSTERS OF IRIS')
plt.xlabel
plt.show()
```

```python
z=pd.Series(y).value_counts()
plt.figure(figsize=(10,8))
plt.pie(z,labels=z.index,autopct='%1.1f%%',shadow=True,startangle=100)
plt.title('PERCENTS OF IRIS')
plt.show()
```

```python
a=pd.Series(y).value_counts().sort_index()
plt.figure(figsize=(10,8))
sns.barplot(x=a.index,y=a.values)
plt.title('NUMBER OF DATAPOINTS IN EACH CLUSTER')
plt.xlabel('CLUSTER LABEL')
plt.ylabel('NUMBER OF DATAPOINTS')
plt.show()
```
