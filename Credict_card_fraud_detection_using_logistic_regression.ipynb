```python
le=LabelEncoder()
creditdf['location_new']=le.fit_transform(creditdf['Location'])
```

```python
le1=LabelEncoder()
creditdf['transaction_type_new']=le1.fit_transform(creditdf['TransactionType'])
```

```python
ind=creditdf[['Amount','MerchantID','TransactionType','Location']]
dep=creditdf['IsFraud']
```

```python
logr=LogisticRegression()
ind=creditdf[['Amount','MerchantID','transaction_type_new','location_new']]
logr.fit(ind,dep)
```

```python
a=float(input('Enter the amount:'))
m=int(input('Enter the merchant id:'))
t=int(input('Enter the transaction type (for purchase 0 and refund 1):'))
l=int(input('Enter the location in integers :'))
answer=logr.predict([[a,m,t,l]])
if answer==1:
  print('Fraud')
else:
  print('Not Fraud')
```
