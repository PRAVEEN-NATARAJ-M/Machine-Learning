```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
customerdf=pd.read_csv("/content/Mall_Customers.csv")
```

```python
customerdf.shape
```

```python
customerdf.info()
```

```python
customerdf.isnull().sum()
```

```python
x=customerdf.iloc[:,[3,4]].values
```

```python
print(x)
```

```python
WCSS=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  WCSS.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),WCSS)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y=kmeans.fit_predict(x)
print(y)
```

```python
plt.scatter(x[y==0,0],x[y==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(x[y==1,0],x[y==1,1],s=100,c='green',label='Cluster 2')
plt.scatter(x[y==2,0],x[y==2,1],s=100,c='purple',label='cluster 3')
plt.scatter(x[y==3,0],x[y==3,1],s=100,c='blue',label='Cluster 4')
plt.scatter(x[y==4,0],x[y==4,1],s=100,c='yellow',label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='black',label='Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
