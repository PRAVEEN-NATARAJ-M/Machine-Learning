```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
```

```python
diadf=pd.read_excel("/content/diabetes.xlsx")
```

```python
diadf.info()
```

```python
diadf.isnull().sum()
```

```python
ind=diadf[['age','mass','insu','plas']]
dep=diadf['class']
```

```python
logr=LogisticRegression()
logr.fit(ind,dep)
```

```python
a=int(input("enter age"))
b=int(input("enter mass"))
c=int(input("enter insu"))
d=int(input("enter plas"))
answer=logr.predict([[a,b,c,d]])
print("RESULT",answer)
```

```python
logr.score(ind,dep)
```

```python
from sklearn.metrics import accuracy_score
```

```python
pval=logr.predict(ind)
accuracy_score(dep,pval)
```
