```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
salp.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
salp['Gen_e']=le.fit_transform(salp['Gender'])
```

```python
le1=LabelEncoder()
salp['edu_e']=le1.fit_transform(salp['Education Level'])
```

```python
salp.isnull().sum()
```

```python
ind=salp[['Age','Gen_e','edu_e','Years of Experience']]
```

```python
dep=salp['Salary']
```

```python
lr=LinearRegression()
```

```python
lr.fit(ind,dep)
```
